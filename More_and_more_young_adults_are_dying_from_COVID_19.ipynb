```python
import pandas as pd

url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_872096a93e4b454da8936d429b20e4fc.csv"
df=pd.read_csv(url)
df.head()
```

```python
df1=df.fillna(0)
df1=df1[df1["Month"]!=0]
df1=df1[df1["Year"]!=0]
df1=df1[df1["State"]=="United States"]
df1=df1[df1["Sex"]=="All Sexes"]
df1=df1[df1["Age Group"]!="All Ages"]
df1=df1[df1["COVID-19 Deaths"]!=0]
```

```python
df1=df1.reset_index(drop=True)
```

##### Apparently, two different methods were used to create the age groups and that resulted in some age groups overlap, for example, "0-17 years" and "1-4 years", "5-14 years". To avoid double counting, we need to select either "0-17 years" or "1-4 years", "5 - 14 years". We will use "0-17 years", "18-29 years", "30-39 years", "40-49 years", "50-64 years", "65 - 74 years", "75- 84 years" and "85 years and over". In other words, we will drop the rows for "Under 1 year", "1-4 years", "5-14 years", "15-24 years", "25-34 years", "35-44 years", "45-54 years", "55-64 years".

```python
df1=df1[df1["Age Group"]!='Under 1 year']
df1=df1[df1["Age Group"]!='1-4 years']
df1=df1[df1["Age Group"]!='5-14 years']
df1=df1[df1["Age Group"]!='15-24 years']
df1=df1[df1["Age Group"]!='25-34 years']
df1=df1[df1["Age Group"]!='35-44 years']
df1=df1[df1["Age Group"]!='45-54 years']
df1=df1[df1["Age Group"]!='55-64 years']
```

```python
summary1=pd.DataFrame(df1.groupby(["Year","Month","Age Group"])["COVID-19 Deaths"].mean()).reset_index()
summary1
```

```python
summary1["Period"]=summary1["Year"].astype(int).astype(str)+"-"+summary1["Month"].astype(int).astype(str)
summary1
```

```python
import plotly.express as px
fig = px.bar(summary1, x="Period", y="COVID-19 Deaths",
             color='Age Group',
             title="COVID-19 deaths in the U.S. by age group.",width=1500)
fig.show()
```

```python
periods=['2020-1','2020-2', '2020-3','2020-4', '2020-5', '2020-6', '2020-7', '2020-8', '2020-9','2020-10', '2020-11', '2020-12',
         '2021-1', '2021-2', '2021-3', '2021-4', '2021-5', '2021-6','2021-7']
for p in periods:
  period=summary1[summary1["Period"]==p]
  fig = px.pie(period, values='COVID-19 Deaths', names='Age Group', title=f'COVID-19 Deaths in the U.S. by age group for the period {p}')
  fig.show()
```
