From March 2007 to March 2021, there were almost 1000 episodes of drought . So, I had to download the drought data in two sets and merge them.

```python
# Merge the drought datasets.
df1=pd.concat([df1a,df1b],axis=0).reset_index(drop=True)
df1.head()
```

```python
# Merge the wildfire dataset
df2=pd.concat([df2a,df2b],axis=0).reset_index(drop=True)
df2.head()
```

##### As shown, the BEGIN_DATE and END_DATE columns for all three datasets are of type "object". We need to convert that into "datetime64[ns]" format so that we can perform mathematical operations on it to get the duration.

```python
def conversion(df):
  df['BEGIN_DATE']= pd.to_datetime(df['BEGIN_DATE'])
  df['END_DATE']= pd.to_datetime(df['END_DATE'])
  return df
```

```python
df1=conversion(df1)
df1.info()
```

```python
df2=conversion(df2)
df2.info()
```

```python
# conversion for the exxcessive heat dataset
df3=conversion(df3)
df3.info()
```

##### The "BEGIN_DATE" and "END_DATE" columns have been converted into "datetime64[ns]" format and we can perform mathematical operations on these columns to get the durations drought.

```python
import numpy as np
def duration(df):
  df["DURATION"]=df['END_DATE']-df['BEGIN_DATE']
  df["DURATION"]=(df['DURATION'] / np.timedelta64(1, 'h'))- df["BEGIN_TIME"]/100+df["END_TIME"]/100
  return df
```

```python
# Duration for the drought episodes.
df1=duration(df1)
df1.head()
```

```python
# Duration of wildfire episodes.
df2=duration(df2)
df2.head()
```

```python
# Duration of the excessive heat episodes
df3=duration(df3)
df3.head()
```

```python
import datetime

def Year(df):
  df["YEAR"]=pd.DatetimeIndex(df["BEGIN_DATE"]).year
  return df
```

```python
# Extract the year from the drought dataset
df1=Year(df1)
df1.head()
```

```python
# Extract year from the wildfire dataset.
df2=Year(df2)
df2.head()
```

```python
# Extract year from the excessive heat dataset.
df3=Year(df3)
df3.head()
```

```python
def Month(df):
  df["Month"]=pd.DatetimeIndex(df["BEGIN_DATE"]).month
  return df
```

```python
# Extract the month from the excessive heat dataset
df3=Month(df3)
df3.head()
```

##### According to the record of the [CALIFORNIA DEPARTMENT OF FORESTRY AND FIRE PROTECTION (Cal Fire)](https://www.fire.ca.gov/media/4jandlhh/top20_acres.pdf), five of six largest wildfires ever recorded in California's history occured in 2020. Among those five largest wildfires, four were recorded in August 2020 and one was recorded in September 2020. What happened in August and September of 2020 in terms of weather?

```python
def summary1(df):
  summary1=pd.DataFrame(df.groupby("YEAR")["EVENT_ID"].count()).reset_index().rename(columns={"EVENT_ID":"#_OF_EPISODES"})
  return summary1
```

```python
# Number of drought episdoes in each year.
summary1_df1=summary1(df1)
summary1_df1
```

```python
# number of wildfire episodes each year.
summary1_df2=summary1(df2)
summary1_df2
```

```python
# number of excessive heat episodes each year.
summary1_df3=summary1(df3)
summary1_df3
```

```python
def summary2(df):
  summary2=pd.DataFrame(df.groupby("YEAR")["DURATION"].mean()).rename(columns={"DURATION":"AVERAGE_DURATION_IN_HRS"}).reset_index()
  return summary2
```

```python
# Average duration of drought episodes each year.
summary2_df1=summary2(df1)
summary2_df1
```

```python
# Average duration of wildfire episodes each year.
summary2_df2=summary2(df2)
summary2_df2
```

```python
# Average duration of excessive heat episodes each year.
summary2_df3=summary2(df3)
summary2_df3
```

##### The analysis shows that 2020 is not a particularly dry years, but it was a particularly hot year. For all the years for which data is available, the year 2020 had the highest number of excessive heat episodes. Furthermore, the average duration of heat episodes was also much longer in 2020. 

##### Let's see how many of those excessive heat episodes in 2020 were recorded in the month of August.
