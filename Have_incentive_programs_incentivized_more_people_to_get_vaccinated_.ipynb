```python
import pandas as pd

#Load the data
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_c5bd3eaccf14437180eb3f730f718de9.csv"
df=pd.read_csv(url,encoding="latin1")
df.head()
```

```python
# Reverse the dataframe so that it has the earlest date on top and latest date at the bottom.
df=df[::-1].reset_index(drop=True)
df.head()
```

##### Define a function to plot the impact of the vaccine incentive programs: Have those programs encouraged more people to get vaccinated? First, calculate the daily increase in percentage. Then, calculate the 7-day moving average of the daily increase in percentage. If the 7-day moving average shows an upward trend after a vaccine incentive program is announced, then, it means the program had incentivize more people to get vaccinated.

```python
import numpy as np
import plotly.express as px

# "s" is the name of the state, all in capital letters and enclosed in quotation marks, i.e., "CALIFORNIA"
df=df[df['Series_Complete_Pop_Pct']!=0].reset_index(drop=True)
states=np.unique(df["State"])
for s in states:
  # Extract the data for a specific state
  state=df[df["State"]==s]
  # Percentage of population that is fully vaccinated.
  daily_pct_increase = []
  # Calculate the daily increase in percentage.
  for i in np.arange(0,len(state)-1):
    dailyinc=state['Series_Complete_Pop_Pct'].iloc[i+1]-state['Series_Complete_Pop_Pct'].iloc[i]
    daily_pct_increase.append(dailyinc)
  daily_pct_increase=pd.DataFrame(daily_pct_increase,columns=["daily_pct_increase"])
  state1=state[['Date','Series_Complete_Pop_Pct']].iloc[1:].reset_index(drop=True)
  state1=pd.concat([state1,daily_pct_increase],axis=1)
  # Calculate the 7-day moving average of the daily increase in percentage.
  state1['7d_rolling_avg_daily_increase(%)']=state1['daily_pct_increase'].rolling(7).mean()
  state1=state1.dropna()
  state1=state1.reset_index(drop=True)
  fig = px.line(state1, x="Date", y="7d_rolling_avg_daily_increase(%)", title=f'Have incentive programs encouraged more people to get vaccinated in {s}?')
  fig.show()
```

##### The charts show that enthusiam for vaccines was high from mid-March to early April. In all states, more and more people got vaccinated everyday. This is reflected on the chart by an upward trend around that time. However, around late April to mid-May, vaccination progress in all states began to slow down.This is reflected by a downward trend from late April to mid-May. 

##### States began to offer incentive programs to incentivize more people to get vaccinated? Have the programs worked as intended? Interestingly, for all states, there was short-period of renewed enthusiasm after announcements of incentive programs. However, those renewed enthusiam lasted for only about 14-days. This bout of enthusiam caused by incentive program is reflected on the graph by a recent upward trend that lasted for about 14 days. However, once the hype for the incentive program was over, ethusiam went back to being lukewarm.

##### So, have the incentives program had worked as incentive? Yes, but only for a short period. The impact of an incentive program is limited. I think it will help if we can find out why people don't get vaccinated.
