```python
import pandas as pd

# Load the data.
url1="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_c5bd3eaccf14437180eb3f730f718de9.csv"
df1=pd.read_csv(url1,encoding = "latin1")

url2="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_229df9e0e5b24090b8985434abf5a83d.csv"
df2=pd.read_csv(url2,encoding = 'latin1')
```

```python
df1.head()
```

```python
# Extract the cumulative statistics for vaccination progress.
df1=df1[::-1]
df3=df1[df1["Date"]=="6/26/2021"]
df3=df3[["Date","Location","State","Administered_Dose1_Pop_Pct","Series_Complete_Pop_Pct"]].sort_values(by=["State"]).reset_index(drop=True)
df3
```

```python
df2.tail()
```

##### The goal is to analyze whether and how the availability of vaccines has impacted COVID-19 deaths. Hence, when we look at death rate, we will look at the data after vaccine has been made available to everyone, not the death rate since the onset of the pandemic. To extract only the data since 5/1/2021, we are going to subtract the statistics on 4/30/2021 from the cumulative data for the latest date. In other words, we are going to use case number and death toll after 5/1/2021.

```python
sub_df2=df2[df2["date"]=="4/30/2021"].sort_values(by=["state"]).reset_index(drop=True)
```

```python
df4=df2[df2["date"]=="6/26/2021"].sort_values(by=["state"]).reset_index(drop=True)
df4["cases"]=df4["cases"]-sub_df2["cases"]
df4["deaths"]=df4["deaths"]-sub_df2["deaths"]
df4
```

##### There is some data cleaning that needs to be done. The extracted data for vaccination progress has 56 rows whereas the dataframe for case number and death toll since 5/1/2021 has only 55 rows. That means there is an extra row in the vaccination progress dataframe. We are going to rename the "Location" column to "code" on the vaccination progress dataf, then merge the two dataframes together using "inner join" and key "code".

```python
df3=df3.rename(columns={"Location":"code"})
df5=pd.merge(df3, df4, on='code', how='inner')
df5
```

##### Drop the redundant columns "date" and "state".

```python
df5=df5.drop(["date","state"],axis=1)
df5.head()
```

##### We are going to define the death rate as the percentage of COVID-19 patients who died. 

```python
df5["death_rate(%)"]=df5["deaths"]*100/df5["cases"]
df5.head()
```

```python
df5=df5.sort_values(by=["Series_Complete_Pop_Pct"],ascending=False)
```

```python
df5=df5.rename(columns={'Administered_Dose1_Pop_Pct':'% of population received 1st dose','Series_Complete_Pop_Pct':"% of population fully vaccinated"})
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def vaccine_impact(col):
  fig = make_subplots(specs=[[{"secondary_y": True}]])
  fig.add_trace(
    go.Bar(x=df5['State'], y=df5[col], name="Percentage of popuplation received the first dose of COVID vaccine"),
    secondary_y=False,)
  fig.add_trace(
    go.Scatter(x=df5['State'], y=df5['death_rate(%)'], name='death_rate(%)'),
    secondary_y=True,)
  fig.update_layout(width=1600)
  fig.update_xaxes(title_text="State")
  fig.update_yaxes(title_text=f"<b>primary</b> {col} ", secondary_y=False)
  fig.update_yaxes(title_text="<b>secondary</b> death_rate(%)", secondary_y=True)
  return fig.show()
```

```python
print("Impact of first vaccine dose on death rate")
vaccine_impact('% of population received 1st dose')
```

```python
print("Impact of full vaccination on death rate")
vaccine_impact('% of population fully vaccinated')
```

##### Death rate is the percentage of COVID patients that died after 5/1/2021. Vaccines had been made available to all people by this time. Did universal eligibility for vaccine help lower death rate? Death rate since 5/1/2021 is below 7% in all states. Every 100 COVID patients, at most 7 people died. However, the graphs show that high percentage of vaccinated population does not correlate to low percentage of deaths among COVID patients. There are some relatively low percentage of fully vaccinated people and low death rate since 5/1/2021, such as Wyoming. In the state of Wyoming, only 34.1% of the population is fully vaccinated, a low percentage compared to states such as Vermont, Massachusetts, Maine, Connecticut, Rhode Island. However, Wyoming also has one of the lowest COVID death rate in the country. Since 5/1/2021, just less than 1% of the COVID patient in the death died. In the state of Maryland, 54.6% of the population are fully vaccinated, a very high percentage. However, Maryland also has the second highest death rate since 5/2/2021. 6.7% of the COVID patients in Maryland died. States with high percentage of fully vaccinated population do not all have low death rates since 5/1/2021. That could be due to the fact that some states had high death rates to begin with; Tt could also be due to the different public health measures each state had.

```python
df2["death_rate(%)"]=df2["deaths"]*100/df2["cases"]

import plotly.express as px
import numpy as np

for s in list(np.unique(df2["state"])):
  df2_s=df2[df2["state"]==s]
  fig = px.scatter(df2_s,x="date", y="death_rate(%)",title=f"Percentage of COVID patients died in {s} since the onset of pandemic")
  fig.show()
```


##### Cumulatively for all states, death rate, the percentage of COVID patients who died, plateaued around 3/1/2021, a timeframe that vaccine became available to most people. 

##### Combining the analytical results from above (the bar charts and the scatter plots), we arrive at the following conclusions: (1) vaccine is the major tool in the fight against COVID-19; (2) the fact that after 5/1/2021, a timeframe when vaccine has been made available to all people, some states with high percentage of fully vaccinated population still have high death rates (as compared to other states), or the fact that in some states fully vaccinated people account for less than 40% of the population, tells us that it is not time to relax all the COVID health guidance yet. Guidances such as masks and social distancing are still necessary.
