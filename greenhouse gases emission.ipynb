```python
import pandas as pd

# load the data.
url1="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_8560419952bb4d6b90042a5dff25c738.csv"
df1=pd.read_csv(url1)

url2="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_023670f7c3cf409bb22debdd6670cf46.csv"
df2=pd.read_csv(url2)
```

```python
df1.head()
```

```python
# Drop the last column. It is the same as the first column and is therefore redundant.
df1=df1.drop(["Gas/Source*.1"],axis=1)
```

```python
df2
```

```python
# Rename the first column and drop the last row.
df2=df2.rename(columns={"Emissions by Economic Sector, MMT CO2 eq.":"Economic sectors"})[:-1]
```

```python
# The dataframe has two levels. The four greenhouse gases (CO2,CH4,N2O and Fluorinated Gases) are the main level,
# the items below them are the sublevels. Extract the main levels and visualize them as percentage of total.
df11=df1[df1["Gas/Source*"]=="CO2"]
df12=df1[df1["Gas/Source*"]=="CH4c"]
df13=df1[df1["Gas/Source*"]=="N2Oc"]
df14=df1[df1["Gas/Source*"]=="Fluorinated Gases(HFCs, PFCs, SF6, and NF3)"]
ghg=pd.concat([df11,df12,df13,df14],axis=0).reset_index(drop=True)
ghg
```

```python
import plotly.express as px
for i in [2005,2015,2016,2017,2018,2019]:
  fig = px.pie(ghg, values=f"{i}",names="Gas/Source*",title=f"Greenhouse gases emission in {i} ")
  fig.show()
```

##### Each year, carbon dioxide CO2 is by far the largest source of greenhouse gases. 

##### Let's see what the biggest contributor of CO2 emission was.

```python
# The contributors to CO2 emissions are the sub-level items listed under CO2 and above CH4.
# Get the index numbers for CO2 and CH4 and then extract the rows between these two index numbers.
CO2=list(df1[df1["Gas/Source*"]=="CO2"].index)
CH4c=list(df1[df1["Gas/Source*"]=="CH4c"].index)
```

```python
df15=df1.iloc[CO2[0]+1:CH4c[0],:]
df15
```

```python
# Visualize the what contributed to CO2 emission and by how much.
for i in [2005,2015,2016,2017,2018,2019]:
  fig = px.bar(df15, x="Gas/Source*", y=f"{i}",
             color='Gas/Source*',height=800,title=f"CO2 emission by sources, {i}")
  fig.update_layout(showlegend=False)
  fig.show()
```

##### The top three sources for CO2 are: Fossil fuel combustion, Transportation, Electric power sector. 

```python
# Visualize greenhouse gases emission by economic sector.
import numpy as np
for i in np.arange(1990,2020):
  fig = px.pie(df2, values=f"{i}",names="Economic sectors",title=f"Greenhouse gases emission by economic sectors, {i} ")
  fig.show()
```

##### The economic sectors that contributed the most to greenhouse gases emission were: Electric generation, Transportation and Industry.

##### What we found from the analyses above:


*   The top source of greenhouse gases was CO2. For each of the years for which data were available, over 80% of the greenhouse gases released into the atmosphere was CO2.
*   The top sources of CO2 were fossil fuel combustion, transportation and electric power sector. Since over 80% of the greenhouse gases was CO2, fossil fuel combustion, transportation and electric power were not only the main sources of CO2, but also the main sources of greenhouse gases emission.
*   The economic sectors that contributed the most to greenhouse gases emission were electric power generation, transportation, Industry.


##### Those findings lead us to this insight: the most impactful approach to reduce greenhouse gases emission would be to reduce CO2 emission and the most impactful approaches to reduce CO2 emission would be to use more clean energy and less fossil fuel, drive more electric vehicles.


##### From [an earlier analysis](https://xiaohuang547.wixsite.com/website-1/post/how-much-co2-does-the-u-s-release-into-the-atmosphere-each-year-will-we-ever-be-net-zero), net emission in the U.S. decreased by only 13.05% from 2005 to 2019.
